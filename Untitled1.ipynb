```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")

print("Path to dataset files:", path)

import pandas as pd
import numpy as np
import os

# Assuming the CSV file is named 'creditcard.csv' and is located inside the downloaded directory
# You may need to adjust the filename if it's different.
file_path = os.path.join(path, 'creditcard.csv')

# Use pd.read_csv to read the CSV file
df = pd.read_csv(file_path)
```

```python
df = df.drop_duplicates()
print("Remaining rows after removing duplicates:", df.shape[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the number of columns
num_cols = df.shape[1]

# Set up the figure
fig, axes = plt.subplots(nrows=num_cols, figsize=(8, 2*num_cols))

# Loop through each column and plot a boxplot
for i, col in enumerate(df.columns):
    sns.boxplot(x=df[col], ax=axes[i])
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df_scaled = scaler.fit_transform(df)
```

```python
plt.figure(figsize=(10,6))
# Convert the scaled NumPy array back to a Pandas DataFrame
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
#Now you can call .corr()
sns.heatmap(df_scaled.corr(),annot=True,cmap='coolwarm')
plt.title("correlation matrix")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_scaled.corr(), annot=False, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap (Improved)")
plt.show()
```

```python
X = df_scaled.drop('Class',axis=1)
y= df['Class']
```

```python
sns.countplot(x=y)
plt.title("before balancing class imbalance")
plt.show()

print("before balancing class imbalance")
print(y.value_counts())
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_sample, y_sample = smote.fit_resample(X,y)

sns.countplot(x=y_sample)
plt.title("after balancing the class imbalance")
plt.show()
print("after balancing the class imablance")
print(y_sample.value_counts())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.2, stratify=y_sample, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=50,         # fewer trees
    max_depth=10,            # shallower trees
    class_weight='balanced',
    n_jobs=-1,               # use all available cores
    random_state=42
)
model.fit(X_train, y_train)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(scale_pos_weight=(y_train.value_counts()[0] / y_train.value_counts()[1]), random_state=42)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))

auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
print("AUC-ROC Score:", auc)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

feature_importance = model.feature_importances_
features = X_train.columns

plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importance, y=features)
plt.title("Feature Importance")
plt.show()
```

```python
top_features = ['V14', 'V13', 'V4', 'V3', 'V8']  # Example
X_train_selected = X_train[top_features]
X_test_selected = X_test[top_features]

# Retrain model
model.fit(X_train_selected, y_train)
```

```python
import shap

# Take a smaller sample of test data
X_sample = X_test_selected.sample(1000, random_state=42)  # Use selected features

# Use TreeExplainer with XGBoost Booster
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_sample)

# Plot the summary
shap.summary_plot(shap_values, X_sample)

```

```python
explainer = shap.TreeExplainer(model.get_booster())
```

```python
# Train new model with selected features
model_selected = XGBClassifier(random_state=42)
model_selected.fit(X_train_selected, y_train)

# Evaluate performance
y_pred_selected = model_selected.predict(X_test_selected)
print(classification_report(y_test, y_pred_selected))
```
